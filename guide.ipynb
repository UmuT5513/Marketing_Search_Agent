# Tavily (web search) API

```python
import os
from dotenv import load_dotenv
import asyncio
from typing import List, TypedDict
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

from tavily import AsyncTavilyClient

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")



# LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)
```

```python


# Tavily
def search_tavily(query):
    query = "World conflict"
    tavily_client = TavilyClient(api_key=TAVILY_API_KEY)
    response = tavily_client.search(query, max_results=3, language="en", start_date="2026-01-01")
    URLs = []
    docs = []
    for item in response["results"]:
        URLs.append(item['url'])
        docs.append(item['content'])
    full_context = "\n\n".join(docs)

    return full_context, URLs, docs

full_context = search_tavily("World conflict")

full_context
```
